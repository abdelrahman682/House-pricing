```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data = pd.read_csv('task.csv')
data = data.dropna()
data.info()
```

```python
x = data.drop(['date', 'street', 'city', 'statezip', 'country', 'price'], axis=1)
y = data['price']
my_model = LinearRegression()
my_model.fit(x,y)
my_model.score(x, y)
```

```python
data = data[data['price']!=0]
```

```python
data = data[data['bedrooms']!=0]
data = data[data['bathrooms']!=0]
data[data==0].count()
```

```python
data.describe()
```

```python
plt.figure(figsize=((15, 6)))
sns.histplot(data['price'], kde=True)
```

```python
data = data[data['price']<.5e07]
plt.figure(figsize=((15, 6)))
sns.histplot(data['price'], kde=True)
```

```python
data['sqft_living'] = np.where(data['sqft_living']>6000, 6000, data['sqft_living'])
plt.figure(figsize=((15, 6)))
sns.scatterplot(x=data['sqft_living'], y=data['price'])
```

```python
data['sqft_above'] = np.where(data['sqft_above']>5000, 5000, data['sqft_above'])
plt.figure(figsize=((15, 6)))
sns.scatterplot(x=data['sqft_above'], y=data['price'])
```

```python
data['sqft_basement'] = np.where(data['sqft_basement']>2000, 2000, data['sqft_basement'])
plt.figure(figsize=((15, 6)))
sns.scatterplot(x=data['sqft_basement'], y=data['price'])
```

```python
data['bedrooms'] = np.where(data['bedrooms']>7, 7, data['bedrooms'])
data['bedrooms'].value_counts().sort_index()
```

```python
plt.figure(figsize=((15, 6)))
sns.barplot(x=data['bedrooms'], y=data['price'])
```

```python
plt.figure(figsize=((15, 6)))
sns.barplot(x=data['bathrooms'], y=data['price'])
```

```python
data['bathrooms'] = np.where(data['bathrooms']>5, 5, data['bathrooms'])
data['bathrooms'] = np.where(data['bathrooms']==1.25, 1, data['bathrooms'])
data['bathrooms'] = np.where(data['bathrooms']==4.75, 4.5, data['bathrooms'])
plt.figure(figsize=((15, 6)))
sns.barplot(x=data['bathrooms'], y=data['price'])
```
